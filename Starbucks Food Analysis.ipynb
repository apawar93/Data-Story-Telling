# Relationships & Trends

The next part of this analysis is to the see the trends and relationships between various numerical columns, to see if sugar, calories, caffeine, carbs, protein have any strong correlations.

```python
Nutrition_Facts.dtypes
```

```python
sns.heatmap(Nutrition_Facts.corr(), annot = True)
```

Observations from the correlation map:

1. Calories has a strong positive correlation with Carbs and Sugar

2. Protein has the Strongest positive correlation with Calories

3. Sugar and Protein have the weakest positive correlation with each other.

4. Protein has weak positive correlation with all except Calories

5. Carbs have strong correlations with Calories and Sugar, and a weak correlation with Fat and Protein

6. Fat only has a strong correlation with Calories.

Other correlations worth exploring are the ones between Protein, Iron, Calcium, Vitamin A, Fiber, and Vitamin C. The reason for doing so it, is to see how strong of relation Protein has with the others. Since Protein has a weak to moderately strong correlation with Fat, Carbs, Sugar, and Calories, it is pertinent to explore how strong of a relation Protein has with these 5 as they are similar to each other.
